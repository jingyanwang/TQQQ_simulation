```python
import pandas as pd

def calculate_sliding_window_returns(df, holding_years, columns):
    # Create a dictionary to store median returns for each holding period for each column
    results = {col: [] for col in columns}
    
    # Loop through each holding period in years
    for years in holding_years:
        # Calculate the number of days corresponding to the holding period (assuming 252 trading days per year)
        window_size = years * 252
        
        # Calculate returns for each column using a sliding window
        for col in columns:
            returns = []
            # Loop through the dataframe with the specified window size
            for i in range(len(df) - window_size):
                # Calculate the return over the window period
                start_price = df[col].iloc[i]
                end_price = df[col].iloc[i + window_size]
                return_percent = (end_price - start_price) / start_price 
                returns.append(return_percent)
            
            # Calculate the median return for this holding period
            median_return = pd.Series(returns).median()
            results[col].append(median_return)
    
    # Create a results dataframe with columns as holding periods and rows as NDX and simulated_QLD
    return_df = pd.DataFrame(results, index=[f"{years} Years" for years in holding_years]).T
    return return_df

# Define the holding periods in years
holding_years = [1, 2, 5, 10, 20, 30]

# Specify the columns to calculate returns on
columns = ['NDX', 'simulated_QLD']

# Calculate returns
returns_df = calculate_sliding_window_returns(QLD_simulation, holding_years, columns)

# Display the result
returns_df
```
